### Using list's append with plots...
<h3>Often you don't have the values you want to plot at the beginning of the program.<br>
You have to do some calculation first and save that number to an array <br>
$$y(x) = A e^{\frac{-x^2}{2 \sigma^2}}$$

Let's say we know that A = 10.0 , sigma = 2.0


````python
import matplotlib.pyplot as plt
import numpy as np
xpoints = [] # This is an empty list for now
ypoints = [] # This is an empty list for now

A = 10.0 # Max of the Gaussian
sigma = 2.0 # half width of the Gaussian

for x in np.arange(-10,10,0.01):
    xpoints.append(x)
    ypoints.append(A*np.exp(-x**2 / (2*sigma**2)))

# I like to convert these list into array as soon as possible
print("xpoints was a ", type(xpoints))
xpoints = np.array(xpoints)
print("xpoints is now", type(xpoints))

# Converting the ypoints to array also.
ypoints = np.array(ypoints)

plt.plot( xpoints , ypoints )
plt.show()
````

````python
import matplotlib.pyplot as plt
import numpy as np
# Now that we know how to plot things
# the most important thing is labelling the axes!

xpoints = [] # This is a list for now
ypoints = []

A = 10.0 # Max of the Gaussian
sigma = 2.0 # half width of the Gaussian

for x in np.arange(-10,10,0.1):
    xpoints.append(x)
    ypoints.append(A*np.exp(-x**2 / 2 / sigma**2))

xpoints = np.array(xpoints)
ypoints = np.array(ypoints)
plt.plot(xpoints,ypoints)

# Before you show your plot
plt.xlabel("Electron Radius [$\AA$]")
plt.ylabel("Probability")
plt.title("Probability Distribution of an Electron's Radius")
plt.show()
````

````python
import matplotlib.pyplot as plt
import numpy as np
# When you are plotting multiple curves in the same plot.
# You HAVE to put a Legend of the curves

xpoints = [] # This is a list for now
ypoints = []

A = 10.0 # Max of the Gaussian
sigma = 2.0 # half width of the Gaussian

for x in np.arange(-10,10,0.1):
    xpoints.append(x)
    ypoints.append(A*np.exp(-x**2 / 2 / sigma**2))

xpoints = np.array(xpoints)
ypoints = np.array(ypoints)
plt.plot(xpoints,ypoints, label="Gaussian") # Added a label to my plot

# Before you show your plot
plt.xlabel("Electron Radius [$\AA$]")
plt.ylabel("Probability")
plt.title("Probability Distribution of an Electron's Radius")
 
    
y = 5*np.sin(xpoints)
plt.plot(xpoints,y,"r",label="sin(x)") # Added a label to my plot # "r--" means, I want red color and ---- lines for the graph

plt.legend()
plt.show()
````

## All the previous plots are of y as a function of x.....
### Some variables are related to each other in a non-functional form
### This is where scatter plots are useful

````python
import matplotlib.pyplot as plt
import numpy as np
# Scatter plots takes exactly the same input as line plots
xpoints = [] # This is a list for now
ypoints = []

A = 10.0 # Max of the Gaussian
sigma = 2.0 # half width of the Gaussian

for x in np.arange(-10,10,0.2):
    xpoints.append(x)
    ypoints.append(A*np.exp(-x**2 / 2 / sigma**2))

xpoints = np.array(xpoints)
ypoints = np.array(ypoints)

# Putting "ko" makes it plot black (k), o's at each point
# Default plot puts lines between each point. 
plt.plot(xpoints,ypoints,"k.",label="Gaussian") 

# Before you show your plot
plt.xlabel("Electron Radius [$\\AA$]")
plt.ylabel("Probability")
plt.title("Probability Distribution of an Electron's Radius")
plt.show()
````

## Example 3.1: Interference of Waves
<h4> When you superimpose two waves they can constructively or destructively interfere with each other.

````python
# Example in 1D
import matplotlib.pyplot as plt
import numpy as np

# the time for the plot
t = np.arange(0,np.pi, 0.01)
omega1 = 6.0
y1 = 5.0*np.sin(omega1*t)
omega2 = 120.0
y2 = 5.0*np.sin(omega2*t)

y12=y1+y2

plt.figure(figsize=(15,5))
plt.plot(t,y2,'-b', label="y2")
plt.plot(t,y1,'-r',label="y1")
plt.legend()
plt.show()


plt.figure(figsize=(15,5))
plt.plot(t,y12, label="y12")
plt.title("y1+y2 intereference")
plt.show()
````

## Histograms: (Not in book)
One of the most useful type of graph in many fields of Physics is the histogram<br>
A histogram keeps track of how often something happen at a particular value<br>
Example of this is students grade histogram (bell curve)<br>

````python
import matplotlib.pyplot as plt
import numpy as np

# A list of students' grades in an exam.
grades = [84,57,69,78,81,78,87,69,90,72,42,81,78,72,45,69,63,
          66,87,75,54,84,84,84,63,90,51,54,78,93,54,69,87,48,
          42,57,78,69,72,66,42,96,63,105,78,69]

plt.plot(grades)
plt.xlabel("Student ID")
plt.ylabel("Grade [%]")
plt.show()

# This is useful if I want to know individual student's grade.
# However, it doesn't tell me overall how the class is doing.
````

````python
# This is where histogram comes in useful.
import matplotlib.pyplot as plt
import numpy as np

# A list of students' grades in an exam.
grades = [84,57,69,78,81,78,87,69,90,72,42,81,78,72,45,69,63,
          66,87,75,54,84,84,84,63,90,51,54,78,93,54,69,87,48,
          42,57,78,69,72,66,42,96,63,105,78,69]


#plt.hist(grades) # Automatic binning
plt.hist(grades, bins=np.arange(40,105,5)) # Choosing my own binning
#plt.xlabel("Grade [%]")
#plt.ylabel("# Students")
#plt.title("Histogram of Student Grades")
plt.show()

# A histogram, the values of the students grades are on the horizontal.
# The number of students that got the grade within the bin is on the vertical.
# We can see that most of the students got 70-75%. 
# No one got below 42%, 
# but there's a second population of students with a lower score than most.
````
