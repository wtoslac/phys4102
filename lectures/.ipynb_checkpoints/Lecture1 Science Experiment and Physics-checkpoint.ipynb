* Question: Are neutrinos its own antiparticle? (Majorana versus Fermionic Neutrino)<br>
    * Antiparticle are the same particle but with an opposite electric charge. Such as a proton (UUD) and antiproton <span style="text-decoration:overline">UUD</span>.

# Experiment Versus Theoretical<br>

Some sciences have a great divide between Experiment and Theory, while other sciences do not.

### For example, in Atmospheric Physics one scientist would be responsible for the entire scientific method:
* The effects of traffic pattern in the Austin, TX area on Nitroxide vertical distribution.
* The daily cycle of morning and evening traffic would have an increase of Nitroxide concentration which distributes upward in the troposphere.
* Obtain data from both ground-based and satelite data to measure the vertical distribution of Nitroxide in the Austin, TX area during multiple time of day.
* Make fits to the vertical distributions with physical parameters to quantify sources and sinks of Nitroxide.
* The data shows that morning traffic creates a high concentration of Nitroxide in the Austin, TX that diffuses upward into the troposphere witha decay constant of 2.4 hours. However, even traffic create a high concentration of Nitroxide that tends to lingers with a decay constant of 8.5 hours due to weaker evening wind pattern.

### In Particle or High Energy Physics the divide tends to be much greater:

For example, 

#### Theoretical particle Physicist would:
* What could be the make-up of dark matter beyond what has been experimentally excluded so far?
* Come up with a new possible interaction that dark matter could be created in the early universe.
* Run a model of the universe to determine how this new interaction could distribute dark matter
* Publish this information and present this new model at conferences.

#### Experimental particle Physicists would:
* Read the theory paper on this new model of dark matter.
* Look at what would the signal that could be detected in a detector.
* Either design a new detector or use data from a current detector to determine how much signal and background this model will have.
* Run statistical analysis on the data to determinte whether or not this new model has a significant signature in the data.
* Publish and present whether or not this new model of dark matter is found.
