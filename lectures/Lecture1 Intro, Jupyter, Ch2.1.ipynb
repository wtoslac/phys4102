## Ch 2.2 Doing Math on Python

````python
# Last time we learn how to declare variables and print them out.
# We can put complicated instruction (program) into the print command
x1 = 4 
x2 = 6
answer= x1+x2

print ("four plus six equals to", answer)
#print ("four plus six equals to", x1+x2)

# Let's examine what happened here.
# Line3: You instructed the computer to create or overwrite the information stored in x1
# Line4: Same for x2
# Line8: print () tells the computer to send whatever is between the () to the screen.
# it reads "4 plus 6 equal to" recognizes that is a string so it will print the characters out one at a time.
# it read x1+x2 and recognize x1 and x2 are int and + is an operator for int.
# so it does the operation first, get a temporary unamed int variable and print out the unamed variable.
````

````python
# There are your typical arithmetics
x3 = x1+x2
print(x3)
x4 = x1-x2
print(x4)
x5 = x1*x2
print(x5)
x6 = x1//x2 # the double slash means integer division
print(x6)
x7 = float(x1)/x2 # a single slash let's python decide if it is floating point division or not.
print(x7)
````

````python
# There's also a way to get the modulus (remainer of the division)
# x1 = 4, x2 = 6
x8 = x1%x2
print(x8)

x8A = x2%x1
print(x8A)
````

````python
# To take a variable to a power use **
x9 = x1**x2 # Assignment operations
print(x9)
print(pow(x1,x2))
````

````python
# The line above x9 = x1**x2 looks like an equation but it is not....
x1**x2 = x9
# a single = sign is an assignment operation
# x9 = x1**x2
# The computer reads x9 should be assigned the value of x1**x2
# So doing it backward will make no sense to the computer. 
````

````python
# If you want to assign multiple variables in one line.
x11 , x12 = x1+x2 , x1*x2 
print (x11)
print (x12)
````

````python
# a double == sign means a comparison operation
a = 10
b = 12
print(a==b)
c = 10
print(a==c)

d = 10+1e-16    # 1 x 10^(-15)
print(a==d)
````
